```python
import numpy as np 
import pandas as pd
```

```python
d = { 'A': [0,np.nan , 2] , 'B' : [0,1,2] , 'C':[2,np.nan , np.nan]}
```

```python
print(d)
```

```python
df = pd.DataFrame(d)
df
```

If dropna() function is depend on axis if axis = 0 then it will work according to row .
If there is any Null or missing value in a row then the whole row will be removed 
When axis = 1 , then it will work according to column . If there is any null or missing value in the 
col then whole col will not consider 

```python
df.dropna()
```

```python
df.dropna(axis=1)
```
