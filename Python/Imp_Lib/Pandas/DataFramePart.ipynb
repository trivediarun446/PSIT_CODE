In pandas there is a frame work name DataFrames . These are used for making tabular data . It is a combination of rows and colume . this frame work offers many function for data manuplation operations

```python
import numpy as np
import pandas as pd 
from numpy.random import randn 
```

```python
tab = pd.DataFrame(randn(5,4), index = 'A B C D E'.split(), columns = ['W','X','Y','Z'])
print(tab)
```

```python
tab['W']
```

```python
tab[['W','X']]
```

```python
type(tab)
```

```python
tab['New'] = tab['W'] + tab['X']
```

```python
print(tab)
```

Using this drop command "New" Column will not delete permanatly .For Deleting Col permanatly we should use impluse attribute . 

```python
tab.drop('New',axis=1)
```

```python
print(tab)
```

```python
tab.drop('New',axis =1 , inplace= True) 
```

```python
print(tab)
```

Conditional Selection 


```python
import numpy as np 
import pandas as pd 
from numpy.random import randn 
```

```python
df = pd.DataFrame(randn(5,5) , index = ' A B C D E'.split() , columns = 'V W X Y Z'.split())
```

```python
print(df)
```

```python
df
```

```python
df['V'] > 0
```

```python
result = df['W']>0
```

```python
result
```

```python
df > 0 
```

```python
df
```

```python
df[df['W']>0]
```

```python
df[df['X']<0]['Z']
```

```python
df[df['W']<0][['Y','Z']]
```

CONDITION => & or | 

```python
df
```

```python
df[(df['W']>0) & (df['Y']>0)]
```

```python
df[(df['W']>0 ) | (df['Y'] >0 )]
```
