```python
import matplotlib.pyplot as plt 
```

```python
fig , axes = plt.subplots()
import numpy as np
x = np.linspace(0,5,11)
y = x**2 
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
# axes= fig.add_axes(0,0,0.1,0.1)
```

```python
fig , axes = plt.subplots(nrows = 1 , ncols = 2) 
for ax in axes :
    ax.plot(x,y,'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('title')
```

a common issue with matplotlib is overlapping subplots or figures. 
    fig.tight_layout() or plt.tight_layout()

```python
fig , axes = plt.subplots(nrows = 3, ncols = 3) 
for ax in axes.flatten():
    ax.plot(x,y,'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('title')
```

```python
fig , axes = plt.subplots(nrows = 3, ncols = 3) 
for ax in axes.flatten():
    ax.plot(x,y,'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('title')
fig.tight_layout()
```


Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch)

```python
fig = plt.figure(figsize=(2,2) , dpi=100)
ax = fig.add_subplot(111)
x = np.linspace(0,5,5)
y= x**2 
ax.plot(x,y,'b')
plt.show()
fig.savefig('newdpi.png',dpi= 200)
```

```python
fig, axes = plt.subplots(figsize=(8,3),dpi = 100)

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
fig.savefig('filename.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'r',label='X square')
ax.plot(x,x**3,'b',label='X cube')
ax.legend(loc=0)
fig.savefig("LEGEND.png")
# we can choose loc from 0 to 10 for position where 0 is for best position 
# according to graph and 10 for center 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'r',label='X square')
ax.plot(x,x**3,'b',label='X cube')
ax.legend(loc=(0.7,0.1))
# we can also change the legend loction according to this attribute 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,lw = 5 ,alpha = 0.5)
# we can use hex code also for color 
# we can use lw attribute for linewidht 
# we can use also transprancy with alpha keyword
# we can use linestyle attribute for styling a line 
```
