```python
import pandas as pd 
```

A Series is very similar to a NumPy array (in fact it is built on top of the NumPy array object). What differentiates the NumPy array from a Series, is that a Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.

In this we learn about series funtion in pandas library 
It is Linear relation data function based on data and index relation 

```python
Company = [ "mahindra" , "tata" , "Kia" , "hyundai" ] 
car = ["Xuv700" , "tiago" , "seltos" , "i20" ] 
ind = ["a", "b" , "c" ,"d " ] 
Number = [1,2,3,4] 
```

```python
data = pd.Series(data = car , index = Company)
print(data)
```

```python
pd.Series(Number , car)
```

```python
pd.Series(car , Number)
```

```python
pd.Series([sum,len,print])
```

```python
pd.Series(Company)
```
